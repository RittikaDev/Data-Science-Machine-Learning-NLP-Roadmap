#### Python Logging

Logging is a crucial aspect of any application, providing a way to track events, errors, and operational information. Python's built-in logging module offers a flexible framework for emitting log messages from Python programs. In this lesson, we will cover the basics of logging, including how to configure logging, log levels, and best practices for using logging in Python applications.

- If you are following this file, and run the second cell first, you have to restart the kernal (project) to see the actual change of the first cell, and vice versa


```python
import logging

## CONFIGURE THE BASIC LOGGING SETTINGS
logging.basicConfig(level=logging.DEBUG)

## LOG MESSAGES WITH DIFFERENT SEVERITY LEVELS
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

#### Log Levels

Python's logging module has several log levels indicating the severity of events. The default levels are:

- DEBUG: Detailed information, typically of interest only when diagnosing problems.
- INFO: Confirmation that things are working as expected.
- WARNING: An indication that something unexpected happened or indicative of some problem in the near future (e.g., ‘disk space low’). The software is still working as expected.
- ERROR: Due to a more serious problem, the software has not been able to perform some function.
- CRITICAL: A very serious error, indicating that the program itself may be unable to continue running.


```python
## CONFIGURING LOGGING
import logging

logging.basicConfig(
    filename="app.log",  # LOG MESSAGES WILL BE WRITTEN TO 'APP.LOG' FILE
    filemode="w",  # 'w' MODE WILL OVERWRITE THE LOG FILE EACH TIME THE PROGRAM RUNS
    level=logging.DEBUG,  # SET THE LOGGING LEVEL TO DEBUG (WILL CAPTURE ALL LEVELS OF LOG MESSAGES)
    format="%(asctime)s-%(name)s-%(levelname)s-%(message)s",  # LOG MESSAGE FORMAT INCLUDES TIMESTAMP, LOGGER NAME, LOG LEVEL, AND THE MESSAGE
    datefmt="%Y-%m-%d %H:%M:%S",  # DATE FORMAT FOR THE TIMESTAMP IN LOG MESSAGES
)

## LOG MESSAGES WITH DIFFERENT SEVERITY LEVELS
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
